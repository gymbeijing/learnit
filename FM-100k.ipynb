```python
data = np.zeros([100000,4])

iloc = 0
with open(os.path.join(dataset_path,'u.data')) as f:
    for line in f:
        data[iloc,:] = line.split('\t')
        iloc += 1
        
ratings = pd.DataFrame(data,columns=['user_id','movie_id','rating','timestamp'])
ratings.head(10)
ratings.describe()
```

```python
genres_info = pd.read_csv(os.path.join(dataset_path,'u.genre'),header=None,delimiter="|")
genres_info = genres_info.rename(columns={0:'genre'}).drop(columns=[1])
genres = list(genres_info['genre'])
print(genres)
```

```python
movies = pd.read_csv(os.path.join(dataset_path,'u.item'),header=None,encoding='ISO-8859-1',delimiter="|")
movies = movies.rename(columns={0:'movie_id',1:'title'}).drop(columns=[2,3,4])
genres_names = {}
for i in range(5,24):
    genre = genres[i-5].lower()
    #genre = re.sub("[^a-zA-Z0-9 ]", " ", genre)
    #genre = genre.strip()
    #genre = re.sub("\s+", " ", genre)
    genres_names[i] = genre
    
movies = movies.rename(columns=genres_names).set_index(['movie_id']).drop(columns=['title'])
#movies.describe()
movies.head(10)
```

```python
from scipy.sparse import csr_matrix
from scipy import sparse

num_ratings = 100000
num_users = 943
num_movies = 1682

# userId matrix with shape[100000,943]
uid_mat = csr_matrix(([], ([], [])), shape=(num_ratings, num_users))

# movieId matrix with shape[100000,1682]
mid_mat = csr_matrix(([], ([], [])), shape=(num_ratings, num_movies))

# uids contains a sequence of user_id in the order of ratings
uids = list(ratings['user_id']-1)
mids = list(ratings['movie_id']-1)

# In each rating, the user who rated the movie is set to one
uid_mat[range(num_ratings), uids] = 1
# In each rating, the movie which is rated is set to one
mid_mat[range(num_ratings), mids] = 1

print(type(uid_mat))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# stop_words = set(stopwords.words('english'))   # have to download
movie_id = list(ratings['movie_id'])
genres = movies.columns
#print(movie_id)

movie_genres = []
for iloc in range(num_movies):
    judge = list(movies.iloc[iloc]==1)
    judge_index = [i for i, x in enumerate(judge) if x]
    movie_genres.append(' '.join([genres[i] for i in judge_index]))

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(movie_genres)
movie_genres_map = X.toarray()
#print(vectorizer.get_feature_names())
print(X.shape)
print(X.toarray()) # X.toarray() is numpy.ndarray
print(movie_genres_map[0])
#print(movie_genres)
    
    
    # find genres of each movie being rated. construct tf-idf of genres
```

```python
genres_mat = np.zeros([num_ratings, 21])

for i in range(num_ratings):
    genres_mat[i] = movie_genres_map[int(mids[i])]

```

```python
genres_mat = sparse.csr_matrix(genres_mat)
print(type(genres_mat))
print(genres_mat.shape)
print(genres_mat)
```

```python
ratings_mat = np.array(list(ratings['rating']))#.reshape((100000,1))
print(ratings_mat)
```

```python
from scipy.sparse import hstack
```

```python
mat = hstack((uid_mat, mid_mat, genres_mat))
```

```python
print(mat.shape)
print(type(mat))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mat, ratings_mat, test_size=0.2)
```

```python
print(X_train.shape)
print(type(X_train))
print(y_train)
```

```python
from pyfm import pylibfm
from sklearn.metrics import mean_squared_error
fm = pylibfm.FM(num_factors=10, num_iter=10, verbose=True, task="regression", initial_learning_rate=0.01,learning_rate_schedule="optimal")
fm.fit(X_train,y_train)
```

```python
y_pred = fm.predict(X_test)
print(y_pred.shape)
```

```python
print(mean_squared_error(y_test, y_pred))
print(y_test[:100])
print(y_pred[:100])
```
